```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/letter-recognition/letter-recognition.data"
cols = ['letter', 'x-box', 'y-box', 'width', 'high', 'onpix', 'x-bar', 'y-bar', 'x2bar', 'y2bar',
        'xybar', 'x2ybr', 'xy2br', 'x-ege', 'xegvy', 'y-ege', 'yegvx']
df = pd.read_csv(url, header=None, names=cols)
```

```python
df
```

```python
X = df.iloc[:, 1:]
y = df['letter']
```

```python
results = []
all_accuracies = []
```

```python
param_dist = {
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'C': np.logspace(-2, 2, 10),
    'gamma': ['scale', 'auto']
}
```

```python
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=i)

    accuracies = []

    random_search = RandomizedSearchCV(
        estimator=SVC(),
        param_distributions=param_dist,
        n_iter=10,
        cv=3,
        verbose=0,
        n_jobs=-1
    )

    random_search.fit(X_train, y_train)
    best_model = random_search.best_estimator_
    best_params = random_search.best_params_

    # Simulate convergence by plotting test score from each candidate
    for candidate in random_search.cv_results_['params']:
        model = SVC(**candidate)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        accuracies.append(acc)

    best_acc = max(accuracies)
    all_accuracies.append(accuracies)

    results.append({
        'Sample': f"S{i+1}",
        'Best Accuracy': round(best_acc * 100, 2),
        'Best Params': best_params
    })
```

```python
df_results = pd.DataFrame(results)
print("\n=== Optimized-SVM Performance Summary ===")
print(df_results.to_string(index=False))
```

```python
best_idx = np.argmax(df_results['Best Accuracy'])
plt.figure(figsize=(8, 5))
plt.plot(all_accuracies[best_idx], marker='o')
plt.title("Convergence (Test Accuracy per Candidate Model)")
plt.xlabel("Candidate Model #")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```
